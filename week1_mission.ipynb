# Misson 2
### Mandatory part
5-MA비트코인 가격 그래프를 그리기

```python
coin=pd.read_csv("C:/cnu/BitCoin.csv")
```

```python
coin
```

```python
coin["Date2"]=[int(i.replace("-","")) for i in coin["Date"]]
coin
```

```python
c=coin[(coin["Date2"]>=20160600) & (coin["Date2"]<=20170699)]
```

```python
c=c.sort_values(by="Date2")
c
```

```python
coin5ma=c["Open"].rolling(window=5).mean()
```

```python
coin5ma2=pd.Series([i for i in coin5ma])

coin5ma2
```

```python
plt.plot(coin5ma2,color="#f2a900")
plt.title("2016.6-2017.6 5ma bitcoin")
plt.xlabel("date")
plt.ylabel("price")
xl=["2016-6","","","2016-9","","","2016-12","","","2017-03","","","2017-06",""]
plt.xticks(np.arange(0,400,30),labels=xl)
plt.show()
```

### Bonus part

```python
ethereum=pd.read_csv("C:/cnu/ETH_day.csv")
ethereum
```

```python
ethereum["Date2"]=[int(i.replace("-","")) for i in ethereum["Date"]]
e=ethereum[(ethereum["Date2"]>=20160600) & (ethereum["Date2"]<=20170699)]
e=e.sort_values(by="Date2")
eth5ma=e["Open"].rolling(window=5).mean()
eth5ma2=pd.Series([i for i in eth5ma])
```

```python
plt.plot(eth5ma2,color="#3c3c3d",label="bitcoin")
plt.plot(coin5ma2,color="#f2a900",label="ethureum")
plt.title("2016.6-2017.6 5ma bitcoin and ethureum")
plt.xlabel("date")
plt.ylabel("price")
xl=["2016-6","","","2016-9","","","2016-12","","","2017-03","","","2017-06",""]
plt.xticks(np.arange(0,400,30),labels=xl)
plt.legend()
plt.show()
```
