# Misson 2
### Mandatory part
5-MA비트코인 가격 그래프를 그리기

```python
coin=pd.read_csv("C:/cnu/BitCoin.csv")
```

```python
# 데이터 확인
coin
```

```python
#날짜를 정수로 바꾼 열을 새로 추가
coin["Date2"]=[int(i.replace("-","")) for i in coin["Date"]]
coin
```

```python
# 20160601~20170630의 데이터만 가져오도록 필터링
c=coin[(coin["Date2"]>=20160600) & (coin["Date2"]<=20170699)]
```

```python
# 날짜를 오름차순으로 정렬
c=c.sort_values(by="Date2")
c
```

```python
# 5일 기준 이동평균
coin5ma=c["Open"].rolling(window=5).mean()
```

```python
# Series를 새로 만들어 index를 다시 부여
coin5ma2=pd.Series([i for i in coin5ma])

coin5ma2
```

```python
plt.plot(coin5ma2,color="#f2a900") # 노란색 선
plt.title("2016.6-2017.6 5ma bitcoin") # 타이틀
plt.xlabel("date") # x축
plt.ylabel("price") # y축
xl=["2016-06","","","2016-09","","","2016-12","","","2017-03","","","2017-06",""] # x축 눈금에 붙일 라벨.
plt.xticks(np.arange(0,400,30),labels=xl) # x축에 새길 눈금은 30일 단위
plt.show()
```
